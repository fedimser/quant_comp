## Visualizing Elementary Cellular Automata

In this Notebook we will visualize all 22 reversible ECA rules, so it becomes clearer what they do.

For every rule we show to pictures. On the left is the evolution of single "enabled" cell. On the right there is evolution of the random state.

```python
from matplotlib import pyplot as plt
import numpy as np
from quantum_eca import BorderCondition, ECA


def single_cell_state(n, pos):
    row0 = np.zeros(n, np.int8)
    row0[pos] = 1
    return row0

def visualize_rules(rules):
    fig, axs = plt.subplots(len(rules), 2, figsize=(15, 7 * len(rules)))
    for i in range(len(rules)):
        ca = ECA(rules[i], BorderCondition.PERIODIC)
        pic0 = ca.visualize(single_cell_state(61, 30), steps=60)
        pic1 = ca.visualize(np.random.choice(2, size=61), steps=60)
        axs[i][0].imshow(pic0, cmap='Greys')
        axs[i][1].imshow(pic1, cmap='Greys')
        axs[i][0].set_ylabel('Rule %d' % rules[i])
    plt.show()

RULES = [15,45,51,60,75,85,89,90,101,102,105,150,153,154,165,166,170,180,195,204,210,240]
visualize_rules(RULES)
```
